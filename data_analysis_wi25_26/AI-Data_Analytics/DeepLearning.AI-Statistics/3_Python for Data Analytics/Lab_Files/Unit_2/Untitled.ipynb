````python
Below is a **complete Jupyter Notebook‚Äìready version**  
‚úî with clean notebook cells (Markdown + Code)  
‚úî plus **exercises + fully worked solutions**  
‚úî covering **correlation, scatterplots, selecting numeric columns**

You can copy‚Äìpaste directly into a `.ipynb` notebook.

---

# üìì **Jupyter Notebook Version ‚Äì Correlation & Scatterplots**

---

## üè∑Ô∏è **Notebook Title:**  
**Exploratory Data Analysis: Correlations & Scatterplots**

---

## üìå **1. Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

---

## üìå **2. Load Dataset**

```python
df = pd.read_csv("your_dataset.csv")
df.head()
```

---

## üìå **3. Basic Scatterplots**

### **Scatterplot 1: Income vs. Money Spent Learning Code**

```python
df.plot(kind='scatter', x='income', y='money_spent_learning_code')
plt.title("Income vs. Money Spent Learning Code")
plt.show()
```

### **Scatterplot 2: Age vs. Months Spent Programming**

```python
df.plot(kind='scatter', x='age', y='months_spent_programming')
plt.title("Age vs. Months Spent Programming")
plt.show()
```

---

## üìå **4. Selecting Numerical Columns**

```python
columns = [
    'age',
    'number_of_children',
    'money_spent_learning_code',
    'months_spent_programming',
    'income'
]

selected_columns = df[columns]
selected_columns.head()
```

---

## üìå **5. Correlation Matrix**

```python
correlation_matrix = selected_columns.corr()
correlation_matrix
```

---

## üìå **6. Optional: Correlation Heatmap (if seaborn allowed)**  
*(If your exam restricts libraries, skip this)*

```python
import seaborn as sns
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

---

# üìù **Exercises + Solutions**

---

## **Exercise 1 ‚Äî Create a scatterplot**

**Task:**  
Create a scatterplot showing the relationship between **income** and **months_spent_programming**.

### ‚úÖ **Solution**

```python
df.plot(kind='scatter', x='income', y='months_spent_programming')
plt.title("Income vs. Months Spent Programming")
plt.show()
```

---

## **Exercise 2 ‚Äî Compute correlation between two variables**

**Task:**  
Calculate the correlation between **age** and **money_spent_learning_code**.

### ‚úÖ **Solution**

```python
df['age'].corr(df['money_spent_learning_code'])
```

---

## **Exercise 3 ‚Äî Create a correlation matrix using only numeric columns**

**Task:**  
Using the defined list `columns`, calculate a correlation matrix.

### ‚úÖ **Solution**

```python
selected_columns = df[columns]
selected_columns.corr()
```

---

## **Exercise 4 ‚Äî Identify the strongest correlation**

**Task:**  
From the correlation matrix, determine the strongest positive correlation.

### ‚úÖ **Solution**

```python
corr_matrix = selected_columns.corr()
corr_matrix.abs().unstack().sort_values(ascending=False).drop_duplicates().head()
```

---

## **Exercise 5 ‚Äî Create two scatterplots and interpret them**

**Task:**  
1. Create a scatterplot: `age` vs `income`.  
2. Create a scatterplot: `months_spent_programming` vs `money_spent_learning_code`.

### üöÄ **Solution**

```python
df.plot(kind='scatter', x='age', y='income')
plt.title("Age vs. Income")
plt.show()

df.plot(kind='scatter', x='months_spent_programming', y='money_spent_learning_code')
plt.title("Programming Months vs. Money Invested")
plt.show()
```

---

# üéØ Want More?
````
