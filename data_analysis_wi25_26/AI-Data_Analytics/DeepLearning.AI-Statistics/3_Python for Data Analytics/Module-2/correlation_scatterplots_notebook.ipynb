# Exploratory Data Analysis: Correlations & Scatterplots

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('your_dataset.csv')
df.head()
```

## Scatterplots

```python
df.plot(kind='scatter', x='income', y='money_spent_learning_code')
plt.title('Income vs. Money Spent Learning Code')
plt.show()
```

```python
df.plot(kind='scatter', x='age', y='months_spent_programming')
plt.title('Age vs. Months Spent Programming')
plt.show()
```

## Selecting Numerical Columns

```python
columns = ['age','number_of_children','money_spent_learning_code','months_spent_programming','income']
selected_columns = df[columns]
selected_columns.head()
```

## Correlation Matrix

```python
selected_columns.corr()
```
