## 3. Adding Interactivity with Widgets

Streamlit supports **widgets** that allow user input:
- `st.button("Click me")`
- `st.slider("Select value", min, max)`
- `st.selectbox("Choose option", options)`

Let's use them to explore the dataset interactively.

More widgets:
- [Input Widgets â†—](https://docs.streamlit.io/develop/api-reference#input-widgets)

# ðŸ“‹

````python
species = st.selectbox("Choose a species:", iris["species"].unique())

filtered_data = iris[iris["species"] == species]

st.write(f"Showing data for: **{species}**")
st.dataframe(filtered_data.head())

# ðŸ§© Student Task 3: Add a slider that allows the user to select between 5â€“30 rows to display from the filtered dataset.
````

## 4. Visualizing Data

You can easily integrate Matplotlib, Seaborn, or Plotly charts in Streamlit. Use `st.pyplot(fig)` or `st.plotly_chart(fig)` to display figures.

More charts:
- [Chart Elements â†—](https://docs.streamlit.io/develop/api-reference#chart-elements)

# ðŸ“‹

````python
import matplotlib.pyplot as plt

# Create a simple scatter plot
fig, ax = plt.subplots()
ax.scatter(iris["sepal_length"], iris["sepal_width"], color="teal")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_title("Sepal Length vs Width")
st.pyplot(fig)

# ðŸ§© Student Task 4: Create a scatter plot for `petal_length` vs `petal_width` instead.
````

## 5. Organization and Structure

Now that we've learnt the basics, let's improve the structure of our app. Streamlit allows you to create clean layouts using Tabs.

Tabs Documentation:
- [Tabs â†—](https://docs.streamlit.io/develop/api-reference/layout/st.tabs)

# ðŸ“‹

````python
tab1, tab2, tab3 = st.tabs(["Cat", "Dog", "Owl"])

with tab1:
    st.header("A cat")
    st.image("https://static.streamlit.io/examples/cat.jpg", width=200)
with tab2:
    st.header("A dog")
    st.image("https://static.streamlit.io/examples/dog.jpg", width=200)
with tab3:
    st.header("An owl")
    st.image("https://static.streamlit.io/examples/owl.jpg", width=200)

# ðŸ§© Student Task 5: Use Tabs to cleanup and organize what we have built so far. Put each section into it's own Tab.
````
