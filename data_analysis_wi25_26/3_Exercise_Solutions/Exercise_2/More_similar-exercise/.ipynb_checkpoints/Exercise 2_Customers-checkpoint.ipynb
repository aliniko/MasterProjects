```python
# Q8. Replace missing values in "TotalCharges" with the median of that column.

df2["TotalCharges"].isna

# median_total = df2["TotalCharges"].median()
# df2["TotalCharges"].fillna(midian_total, inplace = True)

# df["TotalCharges"] = pd.to_numeric(df2["TotalCharges"], errors="coerce").fillna(df2["TotalCharges"].median()).

df2["TotalCharges"] = df2["TotalCharges"].apply(lambda x: x[0] if isinstance(x, (list, np.ndarray)) else x)

df2["TotalCharges"].head()
```

```python
# Q9. Sort the DataFrame by "tenure" in descending order.

sort_tenur_descending = df2.sort_values(by = "tenure", ascending = False)

#sort_tenur_descending
```

```python
# Q10. Create a new column "charge_per_month" as:
# TotalCharges / tenure


#df2.info()


df2['TotalCharges'] = np.floor(pd.to_numeric(df2['TotalCharges'], errors='coerce')).astype('Int64')

charge_per_month = df2.TotalCharges/df2.tenure

#charge_per_month.head()

# print(charge_per_month)

charge_per_month
```
